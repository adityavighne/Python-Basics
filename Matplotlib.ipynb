Matplotlib

**pyplot module**

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots() # fig object is container and ax axis
plt.show()
```

```python
import pandas as pd
df=pd.read_csv('data.csv')
df.head()
```

```python
x=df.Age
y=df.Fare
plt.plot(x,y)

plt.title("First plot")
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
x=df.Age
y=df.Fare
plt.bar(x,y) # bar chart

plt.title("First plot")
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
x=df.Age
y=df.Fare
plt.scatter(x,y)

plt.title("First plot")
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
